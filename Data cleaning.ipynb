## To see histograms of no. of strains vs genes found in them

```python
import pandas as pd
import numpy as np


dataframe=pd.read_excel(r"C:\Users\husay\Desktop\Biotech\New_final_list.xlsx",usecols=[11])
gene_info=dataframe.values.tolist()
num=len(gene_info)
gene_info=np.array(gene_info)
gene_info=np.transpose(gene_info)
print((gene_info))
gene_info=gene_info[0]
```

```python
df=pd.read_excel(r"C:\Users\husay\Desktop\Biotech\New_final_list.xlsx")
print(df.head())
gene_names=df.values.tolist()
print(gene_names[1][0])
```

```python
print(np.max(gene_info)) #ie. there is a gene found in 3766/3771 strains
```

```python
import matplotlib.pyplot as plt

plt.hist(gene_info,bins=5,edgecolor='black')
plt.ylabel("No. of genes")
plt.xlabel("No. of strains gene is found in")
plt.show()

plt.hist(gene_info,bins=100,edgecolor='black')
plt.ylabel("No. of genes")
plt.xlabel("No. of strains gene is found in")
plt.show()
```

```python
index_list=list(np.where(df['No. of Strains']>=3000))
index_list=index_list[0]
final_genes=[]
for index in index_list:
    final_genes.append(df.iloc[index][1]+" ("+df.iloc[index][8]+")")
print(len(final_genes))    #this is the final list of genes we will be analysing
print(final_genes)
```

```python
#deleting the genes we won't be studying
```

```python
import os
from tqdm.notebook import tqdm
list_strains=os.listdir(r"C:\Users\husay\Desktop\Biotech\Species_genomes\Ecoli_strains")
final_strains=[]
for i in tqdm(range(len(final_genes))):
    gene=final_genes[i]
    final_strains.append([])
    for strain in list_strains:
        genes_in_strain=os.listdir(r"C:\Users\husay\Desktop\Biotech\Species_genomes\Ecoli_strains/"+strain)
        if gene in genes_in_strain:
            final_strains[-1].append(strain)
print(len(final_strains))
print(len(final_strains[0]))
```

```python
with open("Gene_list_of_strains.txt" , "w") as file:
    for i in tqdm(range(len(final_strains))):
        file.write("Index="+str(i)+'\n')
        for j in range(len(final_strains[i])):
            file.write(final_strains[i][j]+'\n')
        file.write('\n')
```

```python
union_strains=final_strains[0]
for i in tqdm(range(len(final_strains))):
    union_strains=list(set(union_strains).intersection(final_strains[i]))
print(union_strains)
```
